```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
data = pd.read_csv('anson.csv', names=('e', 'x', 'y', 'z'))
```

```python
data.hist('z', bins=50)
```

```python
cals = data[data.z.abs() < 17.5]
plates = data[data.z.abs() >= 17.5]
```

```python
ax = plt.axes(projection='3d')
ax.scatter(cals.x, cals.y, cals.z)
```

```python
import ipyvolume as ipv
```

```python
c = cals.values.T
ipv.scatter(c[1], c[2], c[3], color='green', cmap=plt.cm.jet)
ipv.show()
```

```python
n = 32

binSize = 35 / n
for i in range(n):
    c = cals[
        (-17.5 + binSize * i <= cals.z) 
        & (cals.z < -17.5 + binSize * (i + 1))
    ]
    fig = plt.figure()
    ax = plt.axes()
    ax.set_xlim(-6, 9)
    ax.set_ylim(-6, 6)
    ax.hist2d(c.x, c.y, cmap=plt.cm.jet, bins=np.linspace(-6, 9, 60))
```
